## Media de temperatura global diaria

En este notebook vamos a visualizar, mediante un gráfico Ridgeline Chart, la distribución de la temperatura mundial entre 1880 y 2014.

El fichero original se puede encontrar en http://berkeleyearth.org/data/

```python
%matplotlib inline

temp = pd.read_csv("daily_temp.csv",comment="%")
temp.head()
```

La columna `Anomaly` contiene la variación de temperatura global diaria en ºC, calculada como la diferencia entre la temperatura media de ese día, y la temperatura global media entre los años 1950 y 1980.

Se puede dibujar la distribución de la temperatura en función del tiempo, agrupando por `Year`, para poder comprobar como la distribución de la temperatura se ha ido desplazando hacia valores más cálidos a lo largo del tiempo.

Para evitar que las etiquetas del eje `y` queden muy juntas y casi imposibles de leer si ponemos todos lo años, primero se crea un lista que pone solo los años múltiplo de 10.

```python
%matplotlib inline

labels=[y if y%10==0 else None for y in list(temp.Year.unique())]
fig, axes = joypy.joyplot(temp, by="Year", column="Anomaly", labels=labels, range_style='own', 
                          grid="y", linewidth=1, legend=False, figsize=(6,5),
                          title="Global daily temperature 1880-2014 \n(°C above 1950-80 average)",
                          colormap=cm.autumn_r)
```
